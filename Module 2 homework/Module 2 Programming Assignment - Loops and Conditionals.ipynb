## 4.1

### **Prompt Answer:**
In this exercise, we assign a hidden number to a variable called `secret` and another number to `guess`. Then, we compare the two values using conditional statements. If `guess` is less than `secret`, the program prints "too low". If `guess` is greater than `secret", it prints "too high". If they are equal, the program prints "just right".


```python
secret = 7
guess = 5

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

## 4.2

### **Prompt Answer:**
This exercise uses Boolean values to describe an object based on two conditions: whether it is small and whether it is green. We assign True or False to both variables, then use conditional statements to determine if the object could be a cherry, pea, watermelon, or pumpkin based on size and color.


```python
small = True
green = True

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

## 6.2

### **Prompt Answer:**
In this exercise, we use a `while` loop to compare a counter (`number`) with a target value (`guess_me`). The loop prints 'too low' while the number is less than `guess_me`. When the values match, it prints 'found it!' and stops the loop. If the number becomes greater than `guess_me`, the program prints 'oops' and exits. The variable `number` increments during each loop cycle.


```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
    
    number += 1
```

## 6.3

### **Prompt Answer:**
This exercise uses a `for` loop instead of a `while` loop to check values from 0 through 9. During each iteration, the loop compares the variable `number` to `guess_me`. If the value is lower, it prints 'too low'. When the number equals `guess_me`, it prints 'found it!' and breaks out of the loop. If the number becomes greater than `guess_me`, it prints 'oops' and exits.


```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
```
