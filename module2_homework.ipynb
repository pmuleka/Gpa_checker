OOP Explanation:
In this exercise, we use a class and a method to sort an array containing 0s, 1s, and 2s. The method counts how many of each value exist, then reconstructs the sorted array. This demonstrates object-oriented programming by organizing the logic inside a class.

```python
class Solution:
    def sort012(self, arr):
        # Count occurrences
        count0 = arr.count(0)
        count1 = arr.count(1)
        count2 = arr.count(2)

        # Build sorted array
        return [0]*count0 + [1]*count1 + [2]*count2


# Example run
sol = Solution()
test_arr = [0, 2, 1, 2, 0, 1]
print("Original:", test_arr)
print("Sorted:", sol.sort012(test_arr))
```

## Functional Programming Exercise: Binary Search

For this task, we use functional programming by implementing binary search using a lambda expression and recursion. The function has no side effects and only returns values, which is a key idea of functional programming. No classes or methods are used, only pure functions.


```python
# Functional Binary Search using lambda and recursion

binary_search = lambda arr, x: (
    None if not arr else
    (len(arr)//2 if arr[len(arr)//2] == x else (
        len(arr)//2 + 1 + (res if (res := binary_search(arr[len(arr)//2+1:], x)) is not None else -len(arr))
        if arr[len(arr)//2] < x else
        (res if (res := binary_search(arr[:len(arr)//2], x)) is not None else None)
    ))
)

# Example run
arr = [1, 2, 3, 4, 5, 6, 7]
search_for = 5
print("Searching for:", search_for)
print("Index found:", binary_search(arr, search_for))
```
